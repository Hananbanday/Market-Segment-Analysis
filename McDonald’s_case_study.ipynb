```python
from sklearn.preprocessing import LabelEncoder
matrix=["yummy", "convenient", "spicy", "fattening", "greasy", "fast", "cheap", "tasty", "expensive", "healthy", "disgusting"] 
for i in matrix:
 df[i]=LabelEncoder().fit_transform(df[i])
df
```

```python
df.mean()
```

```python
x = df.loc[:, matrix]
x
```

```python
x= df.iloc[:,0:11].values
x
```

```python
from sklearn.decomposition import PCA
from sklearn import preprocessing

pcdata = preprocessing.scale(x)

pca = PCA(n_components=11)
pc = pca.fit_transform(x)
names = ['pc1','pc2','pc3','pc4','pc5','pc6','pc7','pc8','pc9','pc10','pc11']
y = pd.DataFrame(data = pc, columns = names)
y
```

```python
var = pca.components_
num_pc = pca.n_features_
pc_list = ["PC"+str(i) for i in list(range(1, num_pc+1))]
var_data = pd.DataFrame.from_dict(dict(zip(pc_list,var)))
var_data['variable'] = x.columns.values
var_data = var_data.set_index('variable')
var_data
```

```python
plt.rcParams
plt.figure(figsize = (12,10))
ax = sns.heatmap(var_data, annot=True)
plt.show()
```

```python
from sklearn.cluster import KMeans

k_values = range(2, 9)
kmeans_results = []
for k in k_values:
    kmeans = KMeans(n_clusters=k, n_init=10)
    kmeans.fit(x)
    kmeans_results.append(kmeans)

def relabel_clusters(cluster_labels):
  for i in range(len(kmeans_results)):
    kmeans_results[i].labels_ = relabel_clusters(kmeans_results[i].labels_)

sum_of_sq_distances = [result.inertia_ for result in kmeans_results]
plt.plot(k_values, sum_of_sq_distances)
plt.xlabel('number of segments')
plt.ylabel('sum of within cluster distances')
plt.show()
```
